## Encoding : 
- **converting categorical to numerical data**
- ML or DL or any models works based on maths.
- models cannot understand the english characters
- so it is very important to convert categorical label to numerical label
- for example gender column has 2 labels male and female , then female represent 0 and male represent 1 based on alpabetical orders.


```python
visa_df['case_status'].unique()
```

```python
# Certifed should assign as 0
# Denied should assign as 1 

d={'Certified' : 0,'Denied' : 1}
d
```

```python
visa_df.case_status=visa_df.case_status.map(d)
```

```python
visa_df.head(3)
```

```python
# continent
labels=sorted(visa_df['continent'].unique())
labels
```

```python
d1={}
for i in range(len(labels)):
    d1[labels[i]]=i
d1
```

```python
visa_df.continent=visa_df.continent.map(d1)
```

```python
visa_df.head()
```

```python
d11={j:i for i,j in enumerate(labels)}   #usin enumerate
d11
```

```python
for k in cat_cols[1:]:
    labels=sorted(visa_df[k].unique())
    d={j:i for i,j in enumerate(labels)}
    visa_df[k]=visa_df[k].map(d)
```

## Label Encoder

- label encoder is a method will conert the categorical variables to numerical variables
- it is under the package **sklearn** (sickit-learn)
- the class name is **preprocessing**
- the method name is **LabelEncoder**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# import the method
# save the method
# apply fit transform
# fit_transform : 
     # - fit : means find the logic   ex: creating mapper or dictionary in above method.

     # - transform : means apply logic and change the data.    ex: apply the mapper(dict) to dataframe(data)
```

```python
visa_df=pd.read_csv(r'C:\Users\RavindharRajula\NareshIT\Data_set_files\Visadataset.csv')
visa_df.head(2)
```
