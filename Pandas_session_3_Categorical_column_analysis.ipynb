```python
#============step-1 create a folder====================
os.makedirs('freq_table',exist_ok=True)

#============step-2 make a new path===================
#path=os.getcwd()+ '\\Frequency_tables\\' + '\\continent_df.csv'
directory=os.getcwd()
folder_name='Frequency_tables'
new_dir=os.path.join(directory,folder_name)

for col in cat_cols[1:]:
    #============step-3 create a DataFrame================
    keys=visa_df[col].value_counts().keys()
    values=visa_df[col].value_counts().values
    df=pd.DataFrame(zip(keys,values),columns=['continent','freq'])
    
    #===========step-4 make the new path==================
    file_name=f'{col}.csv'
    new_path=os.path.join(new_dir,file_name)
    
    #==========step-5 scae into dataframe=================
    df.to_csv(new_path,index=False)
```

```python
#============step-1 create a folder====================
folder_name='Bar_chart_folders'
os.makedirs(folder_name,exist_ok=True)

#============step-2 make a new path===================
#path=os.getcwd()+ '\\Frequency_tables\\' + '\\continent_df.csv'
directory=os.getcwd()
new_dir=os.path.join(directory,folder_name)

for col in cat_cols[1:]:
    #============step-3 create a DataFrame================
    keys=visa_df[col].value_counts().keys()
    values=visa_df[col].value_counts().values
    plt.figure(figsize=(12,6))
    plt.bar(keys,values)
    plt.title(f'{col}_bar_chart')
    plt.xlabel('class')
    plt.ylabel('Freq')
    
    #===========step-4 make the new path==================
    file_name=f'{col}.jpg'
    new_path=os.path.join(new_dir,file_name)
    
    #==========step-5 scae into dataframe=================
    plt.savefig(new_path)
```

```python
#============step-1 create a folder====================
folder_name='Pie_chart_folders'
os.makedirs(folder_name,exist_ok=True)

#============step-2 make a new path===================
#path=os.getcwd()+ '\\Frequency_tables\\' + '\\continent_df.csv'
directory=os.getcwd()
new_dir=os.path.join(directory,folder_name)

for col in cat_cols[1:]:
    #============step-3 create a DataFrame================
    keys=visa_df[col].value_counts().keys()
    values=visa_df[col].value_counts().values
    plt.figure(figsize=(10,5))
    plt.pie(values,labels=keys,autopct='%.1f%%')
    plt.title(f'{col}_bar_chart')
    
    #===========step-4 make the new path==================
    file_name=f'{col}.jpg'
    new_path=os.path.join(new_dir,file_name)
    
    #==========step-5 scae into dataframe=================
    plt.savefig(new_path)
```
