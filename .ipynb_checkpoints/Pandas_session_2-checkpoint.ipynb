**task-1**

- I want to seperate the cat and numerical columns
- cat columns in one list
- Numerical columns in another list.

```python
cols=visa_df.dtypes.keys().tolist()
```

```python
dtype=visa_df.dtypes.values.tolist()
```

```python
cat_lst=[]
numerical_lst=[]
for cat,num in zip(cols,dtype):
    if num=='object':
        cat_lst.append(cat)
    else:
        numerical_lst.append(cat)

```

```python
cat_lst
```

```python
numerical_lst
```

```python
[i for i,j in zip(cols,dtype) if j=='object']  # list comprehension......................
```
